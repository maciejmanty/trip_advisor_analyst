```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"D:\python\projects\learning\trip_advisor_lists")
df.head(15)
```

```python
df = df.dropna()
```

```python
ratings = df["rating"]
new_ratings = []
for x in ratings:
    y = x.split("bubble_")
    y = y[2][:2]
    new_ratings.append(int(y))
df["rating"] = new_ratings

new_ratings = []
for x in df["rating"]:
    y = list(str(x))
    y = ".".join(y)
    new_ratings.append(float(y))
df["rating"] = new_ratings
```

```python
df.head(15)
```

```python
old_prices = df["price"]
new_prices = []
for x in old_prices:
    x = str(x)[4:]
    new_prices.append(x)
df["price"] = new_prices

old_prices = df["price"]
new_prices = []
for x in old_prices:
    x = float(x)
    new_prices.append(x)
df["price"] = new_prices
```

```python
df.head()
```

```python
df.columns = ['names', 'reviews', 'prices', 'ratings']
df.info()
```

```python
reviews = df["reviews"]
new_reviews = []
for x in reviews:
    x = x[:-7]
    x = x.replace(",", "")
    x = x.strip()
    new_reviews.append(int(x))
    
df["reviews"] = new_reviews
```

```python
df.head()
df_reviews = df.sort_values("reviews")
```

```python
sns.lineplot(y=df_reviews["reviews"], x=df_reviews["names"])
```

```python
df_reviews
```

```python
df_reviews_50 = []
def myround(x, base=100):
    return base * round(x/base)

for x in df_reviews["reviews"]:
    x = myround(x)
    df_reviews_50.append(x)
    
```

```python
df_reviews_final = df_reviews
df_reviews_final["reviews"] = df_reviews_50
```

```python
df_to_show = df_reviews_final["reviews"].value_counts()
```

```python
plt.figure(figsize=(16,9))

sns.barplot(y=df_to_show.values, x=df_to_show.index)
```
